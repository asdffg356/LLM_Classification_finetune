```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

typical_path = ROOT / "outputs" / "oof" / "typical_error_cases_for_report.csv"
high_conf_path = ROOT / "outputs" / "oof" / "high_confidence_error_cases.csv"
length_path = ROOT / "outputs" / "logs" / "error_length_analysis.csv"

print("typical exists:", typical_path.exists(), typical_path)
print("high confidence exists:", high_conf_path.exists(), high_conf_path)
print("length analysis exists:", length_path.exists(), length_path)

typical = pd.read_csv(typical_path, encoding="utf-8-sig")
high_conf = pd.read_csv(high_conf_path, encoding="utf-8-sig")
length_analysis = pd.read_csv(length_path, encoding="utf-8-sig")

print("\nTypical columns:")
print(typical.columns.tolist())

print("\nHigh confidence columns:")
print(high_conf.columns.tolist())

print("\nLength analysis:")
display(length_analysis)

def show_existing_columns(df, wanted_cols, n=10):
    existing = [c for c in wanted_cols if c in df.columns]
    print("showing columns:", existing)
    display(df[existing].head(n))

wanted_typical_cols = [
    "case_type",
    "id",
    "label_name",
    "pred_label_name",
    "pred_confidence",
    "true_label_prob",
    "response_len_diff",
    "prompt_short",
    "response_a_short",
    "response_b_short",
    "prompt_clean",
    "response_a_clean",
    "response_b_clean",
]

wanted_high_conf_cols = [
    "id",
    "label_name",
    "pred_label_name",
    "pred_confidence",
    "true_label_prob",
    "response_len_diff",
    "prompt_short",
    "response_a_short",
    "response_b_short",
    "prompt_clean",
    "response_a_clean",
    "response_b_clean",
]

print("\nTypical error cases:")
show_existing_columns(typical, wanted_typical_cols, n=10)

print("\nHigh confidence error cases:")
show_existing_columns(high_conf, wanted_high_conf_cols, n=10)
```
